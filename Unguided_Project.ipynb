```python
# Use this cell to begin your analysis, and add as many as you would like!
import matplotlib.pyplot as plt
import pandas as pd

office_df = pd.read_csv("datasets/office_episodes.csv")
office_df.head()
```

```python
#office_df_ratings = office_df['ratings']
#office_df_ratings.head()
colors = []

for lib, col in office_df.iterrows():
    if col['scaled_ratings'] < 0.25:
        colors.append("red")
    elif col['scaled_ratings'] < 0.50:
        colors.append("orange")
    elif col['scaled_ratings'] < 0.75:
        colors.append("lightgreen")
    else:
        colors.append("darkgreen")
        
print(colors)
```

```python
sizes = []
for lib, col in office_df.iterrows():
    if col['has_guests'] == False:
        sizes.append(25)
    else:
        sizes.append(250)
        
```

```python
plt.rcParams['figure.figsize'] = [11, 7]
fig = plt.figure()
plt.scatter(office_df['episode_number'], office_df['viewership_mil'], color = colors, s = sizes)
plt.title("Popularity, Quality, and Guest Appearances on the Office")
plt.xlabel("Episode Number")
plt.ylabel("Viewership (Millions)")
plt.show()
```
